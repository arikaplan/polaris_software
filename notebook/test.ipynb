```python
fpath = 'C:/Users/akkap/Downloads/Run77.txt' # change this to wherever your file is saved
f = open(fpath,"r")
lines = f.readlines() #read in file row by row
result=[]
for x in lines: 
    row = x.split(' ')[1:] #load each row in split by white spaces
    float_row = [float(i) for i in row] #convert string list elements to floats
    result.append(float_row)
f.close()
```

```python
result = np.array(result) # turn list into array so that you can slice it up in next step
```

```python
index = result[:, 0]
column1 = result[:, 1]
column2 = result[:, 2] #continue like this for each column
```

```python
plt.figure()
plt.plot(index, column1, label='column 1')
plt.plot(index, column2, label='column 2')
plt.xlabel('index')
plt.ylabel('columns')
plt.legend()
```
